# 単回帰モデル

機械学習の学びは、回帰を避けては通れません。

ここでは、最も単純な単回帰モデルについて説明します：

特徴量を用いて、予測値に対する影響の度合いの定数をかけて結果を予測するモデルになります。

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
from sklearn.metrics import mean_squared_error
from sklearn import preprocessing

df= pd.read_csv('./data/winequality-red.csv',sep=';')
# drop aim value
df1 = df.drop(columns='quality')
y = df['quality'].values.reshape(-1, 1)
scaler = preprocessing.StandardScaler()

# Simple Regression
# check the relation between every feature and the result
for column in df1: 
    x = df[column]
    plt.figure() # create a figure
    plt.xlabel(column)
    plt.ylabel('quality')
    plt.scatter(x, y)

    X = x.values.reshape(-1,1)
    X_fit = scaler.fit_transform(X) # scaling
    
    print(f"at first, x was vector: {x.values}")
    print(f"then, convert x to matrix: {X}")
    print(f"finally, scale the matrix: {X_fit}")
    
    model = linear_model.LinearRegression()
    model.fit(X_fit,y)
    plt.plot(x,model.predict(X_fit))
    mse = mean_squared_error(model.predict(X_fit),y)
    print(f"quality = {model.coef_[0][0]} * {column} + {model.intercept_[0]}")
    print(f"MSE: {mse}")
    plt.show()
```

方程式から分かるように、全てのグラフのy切片が同じ。

特徴量がない場合の推定値(x = 0)が、予測対象の `quality` の単純平均値になります。

予測結果に最も正の影響を及ぼす特徴量：alcohol(0.384)

最も負の影響を及ぼす特徴量: volatile acidity(-0.315)

# 重回帰モデル

単回帰モデルは、一つの特徴量に重みをつけて結果を予測するモデルではありますが、

重回帰モデルは、全ての特徴量にそれぞれ重みをつけて結果を予測するモデルであります。

```python
# Multiple regression
# use all features
X = df1.values
X_fit = scaler.fit_transform(X)
model = linear_model.LinearRegression()
model.fit(X_fit,y)
print(model.coef_, model.intercept_)
```

ここでは、やはり切片(intercept)が同じになってます。
